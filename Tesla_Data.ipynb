```python
import pandas as pd

# Load the CSV files (update these paths with your actual file locations)
balance_sheet_df = pd.read_csv('balance_sheet.csv')
cash_flow_df = pd.read_csv('Cash_flow.csv')
income_statement_df = pd.read_csv('Income_statement.csv')
tsla_stock_data_df = pd.read_csv('TSLA.csv')

# Function to convert quarter strings to date format
def convert_quarter_to_date(quarter_str):
    if isinstance(quarter_str, str) and '-' in quarter_str:
        parts = quarter_str.split('-')
        if len(parts) == 2:
            quarter, year = parts
            try:
                year = int(year)
                if quarter == 'Q1':
                    return f'{year}-03-31'
                elif quarter == 'Q2':
                    return f'{year}-06-30'
                elif quarter == 'Q3':
                    return f'{year}-09-30'
                elif quarter == 'Q4':
                    return f'{year}-12-31'
            except ValueError:
                pass
    return None

# Only proceed if the dataframes are not empty
if not balance_sheet_df.empty and not cash_flow_df.empty and not income_statement_df.empty and not tsla_stock_data_df.empty:
    # Apply conversion to the first column of each dataframe and drop rows where conversion fails
    balance_sheet_df['Date'] = balance_sheet_df.iloc[:, 0].apply(convert_quarter_to_date)
    cash_flow_df['Date'] = cash_flow_df.iloc[:, 0].apply(convert_quarter_to_date)
    income_statement_df['Date'] = income_statement_df.iloc[:, 0].apply(convert_quarter_to_date)

    balance_sheet_df = balance_sheet_df.dropna(subset=['Date'])
    cash_flow_df = cash_flow_df.dropna(subset=['Date'])
    income_statement_df = income_statement_df.dropna(subset=['Date'])

    # Convert the 'Date' column in TSLA data to datetime
    tsla_stock_data_df['Date'] = pd.to_datetime(tsla_stock_data_df['Date'])

    # Group TSLA data by quarter and calculate the mean (or sum, depending on the metric)
    tsla_stock_data_df = tsla_stock_data_df.resample('Q', on='Date').mean()

    # Convert the quarterly data to the same format as the financial data
    tsla_stock_data_df.index = tsla_stock_data_df.index.to_series().apply(lambda x: x.strftime('%Y-%m-%d'))

    # Drop the original quarter columns
    balance_sheet_df.drop(balance_sheet_df.columns[0], axis=1, inplace=True)
    cash_flow_df.drop(cash_flow_df.columns[0], axis=1, inplace=True)
    income_statement_df.drop(income_statement_df.columns[0], axis=1, inplace=True)

    # Set 'Date' as the index for merging
    balance_sheet_df.set_index('Date', inplace=True)
    cash_flow_df.set_index('Date', inplace=True)
    income_statement_df.set_index('Date', inplace=True)
    tsla_stock_data_df.index.name = 'Date'

    # Merge the financial dataframes on the 'Date' index
    merged_financial_df = pd.merge(balance_sheet_df, cash_flow_df, left_index=True, right_index=True, how='inner')
    merged_financial_df = pd.merge(merged_financial_df, income_statement_df, left_index=True, right_index=True, how='inner')

    # Merge the financial data with stock data on the 'Date' index
    merged_df = pd.merge(merged_financial_df, tsla_stock_data_df, left_index=True, right_index=True, how='inner')

    # Save the merged dataframe to a new CSV file
    output_file_path = 'Merged_Financial_Stock_Data.csv'
    merged_df.to_csv(output_file_path, index=True)

    print(f'Merged data saved to {output_file_path}')
else:
    print("One or more files are empty. Please ensure the files contain data.")
```
